```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('revOccupiedperSeat (1).xlsx')
df.head()
```

```python
department_data = df.groupby('Department').sum()

# Create a bar chart of total revenue by department
plt.bar(department_data.index, department_data['Revenue'])
plt.title('Total Revenue by Department')
plt.xlabel('Department')
plt.ylabel('Revenue')
plt.show()
```

```python
for department in df['Department'].unique():
    department_data = df[df['Department'] == department]
    plt.plot(department_data['Year'], department_data['Revenue per Seat'], label=department)
plt.title('Revenue per Seat by Year and Department')
plt.xlabel('Year')
plt.ylabel('Revenue per Seat')
plt.legend()
plt.show()
```

```python
for department in df['Department'].unique():
    department_data = df[df['Department'] == department]
    plt.scatter(department_data['Occupancy Ratio'], department_data['Revenue per Seat'], label=department)
plt.title('Occupancy Ratio vs. Revenue per Seat by Department')
plt.xlabel('Occupancy Ratio')
plt.ylabel('Revenue per Seat')
plt.legend()
plt.show()
```

```python
departments = df.groupby('Department')

# Loop through each department and plot the revenue per seat over the years
for name, group in departments:
    plt.plot(group['Year'], group['Revenue per Seat'], label=name)

plt.xlabel('Year')
plt.ylabel('Revenue per Seat')
plt.legend()
plt.show()
```

```python
pivot_df = df.pivot_table(index='Year', columns='Department', values='Revenue', aggfunc='sum')

# Plot a bar chart of the pivot table
ax = sns.barplot(data=pivot_df, palette='Set2')
ax.set_ylabel('Revenue')
ax.set_title('Total Revenue by Department and Year')
plt.show()
```

```python
for name, group in departments:
    plt.scatter(group['Occupied Seats'], group['Revenue'], label=name)

plt.xlabel('Occupied Seats')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

```python
df1= df[df['Year'] == 2021]

# Create scatter plot
plt.scatter(df1['Occupancy Ratio'], df1['Revenue per Seat'])

# Set chart title and axis labels
plt.title('Occupancy Ratio vs. Revenue per Seat for 2021')
plt.xlabel('Occupancy Ratio')
plt.ylabel('Revenue per Seat')

plt.show()
```

```python
pivot_df = df.pivot(index='Year', columns='Department', values='Revenue per Seat')
pivot_df.plot(kind='line')

# Set chart title and axis labels
plt.title('Revenue per Seat by Department')
plt.xlabel('Year')
plt.ylabel('Revenue per Seat')

plt.show()
```

```python
df2 = df[df['Year'] == 2021]

# Pivot data to create a stacked bar chart
pivot_df = df2.pivot(index='Department', columns='Location', values='Revenue')
pivot_df.plot(kind='bar', stacked=True)

# Set chart title and axis labels
plt.title('Revenue by Department and Location for 2021')
plt.xlabel('Department')
plt.ylabel('Revenue (in millions)')

plt.show()
```
